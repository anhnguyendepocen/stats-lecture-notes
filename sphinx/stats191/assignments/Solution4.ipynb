The data set [http://stats191.stanford.edu/data/asthma.table](http://stats191.stanford.edu/data/asthma.table) contains data on the number of admittances, `Y` to an emergency room for asthma-related problems in a hospital for several days. On each day, researchers also recorded the daily high temperature `T`, and the level of some atmospheric pollutants `P`.

- **Fit a linear regression model to the observed counts as a linear function of `T` and `P`.**

```python
%load_ext rmagic
```

```python
%%R
asthma = read.table('http://stats191.stanford.edu/data/asthma.table', header=TRUE)
asthma.lm = lm(Y ~ T + P, data=asthma)
summary(asthma.lm)
```

- **Looking at the usual diagnostics plots, does the constant variance assumption seem justified?**

```python
%%R
par(mfrow=c(2,2))
plot(asthma.lm)
```

There is not very clear evidence of non-constant variance. Perhaps the lower left plot shows some "fan shape" but it is marginal. Nevertheless, we are told below to try weighted least squares.

- **The outcomes are counts, for which a common model is the so-called Poisson model which says that
$\text{Var}(Y) = E(Y)$. In words,  this says that the variance of the outcome is equal to the expected value of the outcome. Using a two-stage procedure, fit a weighted least squares regression to `Y` as a function of `T` and `P` with weights being inversely proportional to the fitted values of the initial model in 1.**

```python
%%R
W = 1/fitted(asthma.lm)
asthma.lm.W = lm(Y ~ T+P, weights=W, data=asthma)
summary(asthma.lm.W)
```

- **Looking at the usual diagnostics plots of this model (which takes the weights into account), does the constant variance assumption seem more reasonable?**

```python
%%R
par(mfrow=c(2,2))
plot(asthma.lm.W)
```

There does not seem to be much difference between the two, either in estimated effects or diagonstic plots. 
The `Scale-Location` plot seems a *little* flatter.

We will perform 1 step of forward stepwise model selection 
when the outcome is generated completely independently from
the design matrix. Following this, we will look at the corresponding $t$-statistics and confidence intervals.

I've written 

```python
%%R
select_first = function(dataset) {
    dataset$Y = rnorm(nrow(dataset))
    first_model = step(lm(Y~1,data=dataset), list(upper=~lcavol + 
                                                  lweight + age + 
                                                  lbph + svi + 
                                                  lcp + pgg45),
                      trace=FALSE, k=0, steps=1)
    return(list(interval=confint(first_model)[2,], pvalue=summary(first_model)$coef[2,4]))
}
select_first(prostate)
```
