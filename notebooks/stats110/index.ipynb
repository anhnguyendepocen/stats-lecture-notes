## Syllabus

### Description

Introduction to statistics for engineers and physical scientists. 

#### Topics

* descriptive statistics
* probability
* interval estimation
* tests of hypotheses
* nonparametric methods
* linear regression
* analysis of variance
* elementary experimental design.
