## Geometric distribution

The geometric distribution describes a model in 
which we wait for a success
to occur in independent trials.

### Probability mass function

We suppose each trial has success probability $q$. The probability that
the first success is at the $j$-th trial is
$$
p(j) = (1-q)^{j-1}q.
$$

This is the *probability mass function*. In the code below $q$ is the variable
`psuccess`.


```python
%%R
psuccess = 0.5
values = seq(0,20)
pmf = dgeom(values, psuccess)
plot(values, pmf, xlab='j', ylab='p(j)')
```

If we change the success probability, the mass function changes.

```python
%%R
psuccess = 0.05
values = seq(0,20)
pmf = dgeom(values, psuccess)
plot(values, pmf, xlab='j', ylab='p(j)')
```

## Distribution function

The CDF (cumulative distribution function) of the geometric with success
probability $p$ is
$$
F(j) = \sum_{1 \leq i \leq j} p(i).
$$


```python
%%R
psuccess = 0.05
values = seq(-1,20)
cdf = pgeom(values, psuccess)
plot(values, cdf, xlab='j', ylab='F(j)', type='s')
```

## Expected value

Let's compute the expected value and variance of a binomial using its mass function.

```python
%%R
psuccess = 0.8
ntrial = 20
vals = seq(0, ntrial)
pmf = dbinom(vals, ntrial, psuccess)
print(sum(pmf))
```

Now that we have the pmf, the mean and variance are simple to compute.

```python
%%R
mean_binom = sum(vals*pmf)
print(c(mean_binom, ntrial*psuccess))
variance_binom= sum((vals-mean_binom)^2*pmf)
print(c(variance_binom, ntrial*psuccess*(1-psuccess)))
```

Here is the variance shortcut the book describes.

```python
%%R
alt_variance = sum(vals^2*pmf) - mean_binom^2
alt_variance
```

## Poisson distribution

We saw in class that the Poisson distribution can arise as a limit
from a Binomial distribution. Specifically, for $\mu > 0$
$$X_n \sim \text{Binom}(n, \mu/n)$$
(which makes sense for any $n > \mu$) we saw that
$$
P(X_n=j) \overset{n \to \infty}{\to} P(Y=j)
$$
where $Y \sim \text{Poisson}(\mu)$.

```python
%%R
n = 20
mu = 3 
print(c(dbinom(0, n, mu/n), dpois(0, mu)))
```

Let's try a little larger $n$:

```python
%%R
n = 200
print(c(dbinom(0, n, mu/n), dpois(0, mu)))
```

```python
%%R
V = seq(0,10,length=11)
binomV = dbinom(V, n, mu/n)
poisV = dpois(V, mu)
plot(binomV, pch=21, ylab='PMF(j)', xlab='j', bg='red', cex=2)
points(poisV, pch=22, bg='blue', cex=2.0)
```

## Example 3.33

* Claim: at most 10% of units needs service at some point in life cycle.

* Take $n=20$ units and see how many eventually need service and compute

$$X = \#  \ \text{units needing service.}$$

* $X \sim Binomial(n, p)$ with $p$ unknown.

* Use the following rule:
   - If $X \geq 5$, more than 10% need service.
   - If $X \leq 4$, less than 10% need service.
   
1. Assume $p=0.1$, what is $P(X \geq 5)?$

2. Assume $p=0.2$ what is $P(X \leq 4)$?

In `R`, the argument `lower.tail=FALSE` computes
$$
P(X > q) = 1 - P(X \leq q) = 1 - F_X(q).
$$

```python
%%R
print(1 - pbinom(4, 20, 0.1))
print(pbinom(4, 20, 0.2))
```

```python
%%R
values = seq(0.1, 1, length=50)
plot(values, pbinom(4, 20, values), ylab='1 - Power')
```

## Example 3.40

Publisher estimates:

* Errors are independent from page to page.

* For a 400 page book.

P(exactly one error)?
