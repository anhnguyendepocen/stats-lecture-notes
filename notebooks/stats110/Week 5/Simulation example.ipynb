## Simulating from a distribution and summarizing results

In problem 5.44, you are asked to carry out a simulation experiment for a
particular distribution, the Weibull distribution. In `R` the relevant functions are `dweibull, pweibull, qweibull, rweibull`.

In this example, we will carry out a similar experiment but for 
Gamma random variables.

We will look at the distribution of the sample mean when drawing `n` samples
from $\Gamma(\alpha,\beta)$, i.e. with density
$$
f_X(x;\alpha,\beta) = \frac{1}{\Gamma(\alpha) \beta^{\alpha}} x^{\alpha-1} e^{-x/\beta}  \qquad x \geq 0
$$

```python
%%R
mysample.mean = function(n, alpha, beta) {
    gamma_sample = rgamma(n, alpha, scale=beta)
    return(mean(gamma_sample))
}

n = 10
alpha = 2
beta = 4
mysample.mean(n, alpha, beta)
```

Let's take $k$ draws from this sample mean distribution
and draw the histogram.

```python
%%R
k = 10000
sample_of_means = c()
for (i in 1:k) {
    sample_of_means = c(sample_of_means, mysample.mean(n, alpha, beta))
}
print(length(sample_of_means))
```

```python
%%R
hist(sample_of_means)
```

A smoother visual summary can be obtained by estimating the density
from a sample. This is called a `kernel density estimate`.

```python
%%R
plot(density(sample_of_means))
```

In fact, from our earlier example, we know what the density is exactly.

```python
%%R
plot(density(sample_of_means), lwd=4, lty=2, col='blue')
X = seq(0,15,length=201)
lines(X, dgamma(X, alpha*n, scale=beta/n), lwd=3, col='red')
```

## Increasing n to 50

Our earlier example took $n=10$ draws and repeated the experiment 10000 times.
Let's instead take a sample of $n=50$ draws.

```python
%%R
n = 50
k = 10000
sample_of_means2 = c()
for (i in 1:k) {
    sample_of_means2 = c(sample_of_means2, mysample.mean(n, alpha, beta))
}
plot(density(sample_of_means2), lty=2, col='blue', lwd=4)
X = seq(5,11,length=201)
lines(X, dgamma(X, alpha*n, scale=beta/n), lwd=3, col='red')
```

## Central Limit Theorem 

Taking more draws made the histogram look more like the normal curve. It also made
the sample mean more concentrated (i.e. the horizontal axis is more compressed).

The Central Limit Theorem says that the average of both sets of samples is 
the expected value of one of the draws: $\alpha \cdot \beta = 2 \cdot 4 = 8$.

```python
%%R
print(c(mean(sample_of_means),mean(sample_of_means2)))
```

But the variance of the two is determined by $n$.
The variance of the first set of samples should be roughly
$$
\alpha \cdot \beta^2 / 10 = \frac{2 \cdot 16}{10} = 3.2
$$

```python
%%R
var(sample_of_means)
```

While for the second set of samples, the variance should be
$$
\alpha \cdot \beta^2 / 50 = \frac{2 \cdot 16}{50} = 0.64
$$

```python
%%R
var(sample_of_means2)
```

Let's look at our Gamma example again.

```python
%%R
plot(density(sample_of_means2), lty=2, col='blue', lwd=4)
X = seq(5,11,length=201)
lines(X, dgamma(X, alpha*n, scale=beta/n), lwd=3, col='red')
lines(X, dnorm(X, mean=alpha*beta, sd=sqrt(alpha*beta^2/n)), lwd=4, lty=3, col='green')
```
