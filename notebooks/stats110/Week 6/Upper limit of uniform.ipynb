## Upper limit of uniform on $[0,\theta]$

Another example we saw in the slides for Chapter 6 is estimation of $\theta$ when
observing a sample 
$$X_1, \dots, X_n \overset{IID}{\sim} \text{Uniform}(0,\theta).$$
That is, the $X_i$'s are independent and each $X_i$ has density
$$
f(x;\theta) = \begin{cases} \frac{1}{\theta} & x \in [0,\theta] \\
0 & \text{otherwise.}
\end{cases}
$$

Two unbiased estimators were proposed:

* The maximum, scaled to be unbiased: $\widehat{\theta}_1 = \frac{n+1}{n} \cdot \max(X_1, \dots, X_n) = \frac{n+1}{n} \cdot X_{(n)}$.

* An estimate based on $\bar{X}$: $\widehat{\theta}_2 = 2 \bar{X}$.


```python
%%R
theta = 2
theta.hat = function(X) {
    n = length(X)
    return(c((n+1)/n*max(X), 2*mean(X)))
}
```

Let's draw samples of these two estimators based on 1000 replicates with a sample size of 10.

```python
%%R
nreplicates = 1000
sample.size = 10
estimators = c()
for (i in 1:nreplicates) {
    estimators = rbind(estimators, theta.hat(runif(sample.size, min=0, max=theta)))
}
plot(density(estimators[,1]), lwd=3, col='green', xlim=c(1,3))
lines(density(estimators[,2]), lwd=3, col='red')
```

Let's increase our sample size to 100 and see if the difference is exaggerated
even more. Based on the size of the variance, the difference *should* grow as the sample 
size grows.

```python
%%R
nreplicates = 1000
sample.size = 100
estimators = c()
for (i in 1:nreplicates) {
    estimators = rbind(estimators, theta.hat(runif(sample.size, min=0, max=theta)))
}
plot(density(estimators[,1]), lwd=3, col='green', xlim=c(1.7,2.3))
lines(density(estimators[,2]), lwd=3, col='red')
```
