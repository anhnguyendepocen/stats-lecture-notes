## Bootstrap estimate of standard error

Another example encountered in Chapter 6 is the *bootstrap estimate* of
standard error. It is based on what statisticians call *resampling*.

The book's example involves estimating the rate $\lambda$ based on a sample of
10 $\text{Exp}(\lambda)$ random variables.

```python
%%R
breakdown = c(41.53, 18.73, 2.99, 30.34, 12.33, 117.52, 73.02, 223.63, 4.00, 26.78)
lambda.hat = 1 / mean(breakdown)
lambda.hat
```

### Bootstrap distribution

The idea the bootstrap exploits is that, if the estimate of $\lambda$ is reasonable, then one can get a sense of the variability in the estimate be *resampling* 10 independent $\text{Exp}(\widehat{\lambda})$ random variables and recomputing
the estimate of $\lambda$.

For the $b$-th bootstrap replicate, $\lambda^*_b$ is the estimate of $\lambda$
from that bootstrap sample. Here is our first bootstrap sample:

```python
%%R
lambda_1 = 1 / mean(rexp(10, rate=lambda.hat))
lambda_1
```

Instead of just the standard error, the bootstrap gives us a pretty good
idea of the sampling distribution of our estimator.

```python
%%R
nreplicate = 1000
bootstrap_samples = c()
for (i in 1:nreplicate) {
    bootstrap_samples = c(bootstrap_samples, 1 / mean(rexp(10, rate=lambda.hat)))
}
plot(density(bootstrap_samples), lwd=3, col='red')
```

We see that the sampling distribution is not quite 
normal -- it has some right skew in it.
We get an estimate of the bias, as well as standard error. 

```python
%%R
bias_estimate = lambda.hat - mean(bootstrap_samples)
sd_estimate = sd(bootstrap_samples)
data.frame(bias_estimate, sd_estimate)
```
