# Estimating the population mean

Chapter 6 on point estimation discusses the problem of estimating the population
mean based on a sample. They use four estimators as motivating examples:

* Sample mean: $\bar{X} = \frac{1}{n} \sum_{i=1}^n X_i$

* Sample median: $\tilde{X} = X_{(n/2)}$ where $X_{(i)}$ are the order statistics
of $(X_1,\dots, X_n)$ (i.e. the ordered values of the sample). If the sample size is even, then the sample median is the midpoint of the two middle sample values...

* The average of the two extreme values: $(X_{(n)}-X_{(1)}) / 2$.

* The trimmed mean: $\bar{X}_{\text{tr}(\alpha)}$ discard $\alpha \cdot 100 \%$  of the highest and lowest order statistics and return the mean of the remaining points.

We will illustrate their properties for a few different sampling distributions.

```python
%%R
mean.estimators = function(X) {
    return(c(mean(X), median(X), 0.5 * (max(X) + min(X)), mean(X, trim=0.1)))
}
mean.estimators(rnorm(50))
```

## Noise models

In all cases, the population mean will be 2, though we will add different types
of noise to 2.

* Normal noise with variance 1: `rnorm(..., mean=0, sd=1)`

* Cauchy noise with scale 1: `rcauchy(..., location=0, scale=1)`

* Uniform noise on [-1,1]: `runif(..., min=-1,max=1)`

Based on samples of different sizes we will look at a density estimate of the
resulting estimator. Tight density estimates indicate an accurate estimator.

```python
%%R
pop.mean = 2
normal.noise = function(n) { return(rnorm(n, mean=0, sd=1) + pop.mean)}
cauchy.noise = function(n) { return(rcauchy(n, location=0, scale=1) + pop.mean)}
uniform.noise = function(n) { return(runif(n, min=-1, max=1) + pop.mean)}

sample.estimator = function(noise.model, sample.size, nreplicates) {
    estimators = c()
    for (i in 1:nreplicates) {
        estimators = rbind(estimators, mean.estimators(noise.model(sample.size)))
    }
    return(estimators)
}
```

### Normal noise, sample size = 20


```python
%%R
normal.sample = sample.estimator(normal.noise, 20, 1000)
plot(density(normal.sample[,1]), col='red', main='Normal noise, sample size=20', lwd=3)
lines(density(normal.sample[,2]), col='blue', lwd=3)
lines(density(normal.sample[,3]), col='green', lwd=3)
lines(density(normal.sample[,4]), col='orange', lwd=3)

```

The Cauchy distribution has very heavy tails. So heavy that the distribution does
not have a variance (and so the Central Limit Theorem does not apply).

The sample mean and average of the extreme value estimators are terrible for this noise model.
