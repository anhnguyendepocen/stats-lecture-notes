## Type I and Type II

In class, we considered two specific hypothesis tests

### Example 8.1 -- Car Damage

Here,
$H_0$ states that 25% of cars have no damage at collision speeds of 10 MPH. 
The alternative, $H_a$ says that there has been an improvement, i.e. more than 25\%
have no damage.

We decided to reject $H_0$ if, in our sample of 20 cars 8 or more showed no damage.

The Type I error rate is:

```python
%%R
1 - pbinom(7, 20, 0.25)
```

The type II error rate depends on the actual percentage in the improved cars.

```python
%%R
values = seq(0.25, 1, length=50)
plot(values, pbinom(7, 20, values), type='l', ylab='Type II error rate', xlab='True proportion, p',ylim=c(0,1))
```

The power is 1 - Type II error rate

```python
%%R
values = seq(0.25, 1, length=50)
plot(values, 1 - pbinom(7, 20, values), type='l', ylim=c(0,1), ylab='Power(p)', xlab='True proportion, p')
```

### Example: nicotine content

Here, $H_0$ was that the average nicotine content was less than 1.5mg per cigarette while $H_a$
states that the average is larger. We assume that the standard deviation of nicotine
content is 0.20mg and that the content is normally distributed.

Our threshold, based on a sample of 32 cigarettes is: 

```python
%%R
threshold = 1.5 + 1.65 * 0.2 / sqrt(32)
threshold
```

```python
%%R
1 - pnorm(1.65)
```

The power again depends on the true average nicotine content.

```python
%%R
values = seq(1.5,1.7,length=50)
plot(values, 1 - pnorm(threshold, mean=values, sd=0.2 / sqrt(32)), xlab='True content (mu)', ylab='Power(mu)', type='l', ylim=c(0,1))
```

We might have performed a two-sided test. In this case, the power looks slightly
different.

```python
%%R
values = seq(1.3, 1.7, length=100)
plot(values, 1 - pnorm(1.5 + 1.96 * 0.2 / sqrt(32), mean=values, sd=0.2 / sqrt(32))
     + pnorm(1.5 - 1.96 * 0.2 / sqrt(32), mean=values, sd=0.2 / sqrt(32)), xlab='True content (mu)', ylab='Power(mu)', type='l', ylim=c(0,1))
```

### Sample size

Suppose we had a different sample size instead of 32 in the nicotine example
and suppose the true mean nicotine content is 1.55 instead of 1.5.

We could ask the question: *how large a sample do we need in order for our
test to have power, say, 80%*.

```python
%%R
truth = 1.55
nvalues = seq(10, 150, length=41)
plot(nvalues, 1 - pnorm(1.5 + 1.65 * 0.2 / sqrt(nvalues), mean=truth, 
                        sd=0.2 / sqrt(nvalues)), xlab='Sample size (n)', ylab='Power(n)', type='l', ylim=c(0,1))
abline(h=0.8, lty=2)
```
